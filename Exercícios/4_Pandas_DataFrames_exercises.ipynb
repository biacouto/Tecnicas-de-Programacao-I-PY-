###Crie um  `marvel_df` com os dados abaixo

```python
marvel_data = [
    ['Spider-Man', 'male', 1962],
    ['Captain America', 'male', 1941],
    ['Wolverine', 'male', 1974],
    ['Iron Man', 'male', 1963],
    ['Thor', 'male', 1963],
    ['Thing', 'male', 1961],
    ['Mister Fantastic', 'male', 1961],
    ['Hulk', 'male', 1962],
    ['Beast', 'male', 1963],
    ['Invisible Woman', 'female', 1961],
    ['Storm', 'female', 1975],
    ['Namor', 'male', 1939],
    ['Hawkeye', 'male', 1964],
    ['Daredevil', 'male', 1964],
    ['Doctor Strange', 'male', 1963],
    ['Hank Pym', 'male', 1962],
    ['Scarlet Witch', 'female', 1964],
    ['Wasp', 'female', 1963],
    ['Black Widow', 'female', 1964],
    ['Vision', 'male', 1968]
]
```

```python
df_marvel = pd.DataFrame(marvel_data)
df_marvel
```


### Adicione a coluna nomes ao DF


```python
df_marvel=pd.DataFrame(marvel_data, columns=['name', 'gender', 'year'])
df_marvel
```

#
##Adicione o index `marvel_df` (use o nome dos personagens como index)

```python
df_marvel['index_marvel'] = df_marvel['name']
df_marvel = df_marvel.set_index('index_marvel')
df_marvel
```

###Drop a coluna nome visto que agora ela  o index

```python
df_marvel = df_marvel.drop('name', axis=1)
df_marvel
```


###Drop Namor e Hank Pyn


```python
df_marvel = df_marvel.drop(labels=['Namor', 'Hank Pym'], axis=0)
df_marvel
```

## DataFrame manipulao e operao

### Modifique a  `first_appearance` do 'Vision' para o ano 1964



```python
df_marvel.loc[df_marvel.index == 'Vision', 'year'] = 1964
df_marvel
```


### Adicione uma nova coluna para o dataframe  `marvel_df` chamada   'years_since' com os anos desde  `first_appearance`


```python
current_year = 2024
df_marvel['years_since'] = current_year - df_marvel['year']
df_marvel

```

```python
df_marvel = df_marvel.rename(columns={'year': 'first_appearance'})
df_marvel
```


## DataFrame boolean arrays

### D ao  `marvel_df` mostrando somente os personagens femininos


```python
female = df_marvel[df_marvel['gender'] == 'female']
female
```


### Given the `marvel_df` pandas DataFrame, get the male characters


```python
male = df_marvel[df_marvel['gender'] == 'male']
male
```


### Given the `marvel_df` pandas DataFrame, get the characters with `first_appearance` after 1970


```python
appearance = df_marvel[df_marvel['first_appearance'] > 1970]
appearance
```




### Given the `marvel_df` pandas DataFrame, get the female characters with `first_appearance` after 1970

```python
female_appearance = df_marvel[(df_marvel['gender'] == 'female') & (df_marvel['first_appearance'] > 1970)]
female_appearance

```
